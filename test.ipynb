```python
import cv2
import numpy as np
```

```python
model = YOLO("C:\\Users\\owend\\Projects\\Eye_tracker\\runs\\detect\\train2\\weights\\best.pt")  # .pt file from training

# Open the webcam (or use a video file)
cap = cv2.VideoCapture(1)  # or 'your_video.mp4'

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO inference on the frame
    results = model(frame)

    # Loop through detections and draw boxes
    for r in results:
        boxes = r.boxes
        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box
            conf = float(box.conf[0])               # Confidence
            cls = int(box.cls[0])                   # Class id
            label = model.names[cls]                # Class name

            # Draw the bounding box and label
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, f"{label} {conf:.2f}", (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    cv2.imshow('YOLOv11 Detection', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
